### Part One: Load Data

```python
import pandas as pd
```

```python
phone = pd.read_csv('cell_phones.csv')
pop = pd.read_csv('population1.csv')
geo = pd.read_csv('ddf--entities--geo--country.csv')
```

### Part Two: Join Data

```python
pop.columns
```

```python
df2 = phone.merge(pop,  on=['Country', 'year'], how='inner')
```

```python
df2
```

```python
geo.columns
```

```python
geo.rename(columns={'country':'geo', 'name':'Country'},inplace=True )
```

```python
geo[['geo', 'Country']]
```

```python
df3 = df2.merge(geo[['geo', 'Country']])
```

```python
df3
```

### Part Three: Make Features

```python
df3['cell_phone_per_person'] =  df3['cell_phones_total']/df3['population']
```
